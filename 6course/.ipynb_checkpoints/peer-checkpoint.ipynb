**Создайте на основе функций *prepare_train_set* и *prepare_sparse_train_set_window*  новую – *prepare_train_set_with_fe*, (от "feature engineering"), создайте следующие признаки:**
- `session_timespan` – продолжительность сессии (разница между максимальным и минимальным временем посещения сайтов в сессии, в секундах)
- `#unique_sites` – число уникальных сайтов в сессии 
- `start_hour` – час начала сессии (то есть час в записи минимального timestamp среди десяти)
- `day_of_week` – день недели (то есть день недели в записи минимального timestamp среди десяти)

Функция должна возвращать новый DataFrame (как возвращала функция *prepare_train_set*), только признаков должно быть на 4 больше. Порядок, в котором добавляются признаки: *site1*, ... *site10*, *session_timespan*, *#unique_sites*, *start_hour*, *day_of_week* и *user_id* (это видно и чуть ниже по тому, как функция вызывается).

```python
def prepare_train_set_with_fe(path_to_csv_files, site_freq_path, session_length=10, window_size=10):
    with open(site_freq_path, 'rb') as site_freq_pkl:
        site_freq = pickle.load(site_freq_pkl)
    res=[]
    for i in tqdm_notebook(glob(path_to_csv_files+'/*.csv'), desc = 'Создаем список сайтов'):
        user = pd.read_csv(i)
        new_list = list(map(lambda x: site_freq[x][0], user.site.values))
        new_list1 = list(map(lambda x: datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S' ), user.timestamp.values))
        for p in np.arange(0,len(new_list), window_size):
            if (p+session_length)<=len(new_list):
                res.append(new_list[p:(p+session_length)])
                res[-1] = np.append(res[-1],(new_list1[p+session_length-1]-new_list1[p]).seconds)
                res[-1] = np.append(res[-1],len(set(new_list[p:(p+session_length)])))
            else:
                res.append(np.append(new_list[p:len(new_list)],np.zeros(p+session_length-len(new_list))))
                res[-1] = np.append(res[-1],(new_list1[len(new_list1)-1]-new_list1[p]).seconds)
                res[-1] = np.append(res[-1],len(set(new_list[p:len(new_list)])))
            res[-1] = np.append(res[-1],new_list1[p].hour)
            res[-1] = np.append(res[-1],new_list1[p].weekday())
            res[-1] = np.append(res[-1],int(i[-8:-4]))
    col = []
    for i in range(session_length):
            col.append('site'+str(i+1))
    col.append('session_timespan')
    col.append('#unique_sites')
    col.append('start_hour')
    col.append('day_of_week')
    col.append('target')
    session_data = pd.DataFrame(res, columns = col,  dtype=int)
    return session_data
```

Сделайте это только для набора данных, полученного с параметрами *session_length=10* и *window_size=10*. Сериализуйте полученные матрицы с помощью pickle. Функция может возвращать как только новые признаки, так и старые с новыми. При этом сигнатура функции может быть другой – тут уже свобода выбора.**

```python
#добавил в функцию новые признаки для сессии: количество посещений топ30 сайтов, время суток, индикатор рабочего дня
#и время года. Из старых оставлены длина сессии и количество разных сайтов в сессии.
def feature_engineering(path_to_csv_files, site_freq_path, session_length=10, window_size=10):
    season_dic = {1:0,2:0,3:1,4:1,5:1,6:2,7:2,8:2,9:3,10:3,11:3,12:0}
    with open(site_freq_path, 'rb') as site_freq_pkl:
        site_freq = pickle.load(site_freq_pkl)
    res=[]
    for i in tqdm_notebook(glob(path_to_csv_files+'/*.csv'), desc = 'Создаем список сайтов'):
        user = pd.read_csv(i)
        new_list = list(map(lambda x: site_freq[x][0], user.site.values))
        new_list1 = list(map(lambda x: datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S' ), user.timestamp.values))
        for p in np.arange(0,len(new_list), window_size):
            if (p+session_length)<=len(new_list):
                res.append(new_list[p:(p+session_length)])
                res[-1] = np.append(res[-1],(new_list1[p+session_length-1]-new_list1[p]).seconds)
                res[-1] = np.append(res[-1],len(set(new_list[p:(p+session_length)])))
            else:
                res.append(np.append(new_list[p:len(new_list)],np.zeros(p+session_length-len(new_list))))
                res[-1] = np.append(res[-1],(new_list1[len(new_list1)-1]-new_list1[p]).seconds)
                res[-1] = np.append(res[-1],len(set(new_list[p:len(new_list)])))
            res[-1] = np.append(res[-1],sum(1 for n in res[-1][:10] if n <=30))#добавил число посещений топ-30 сайтов
            #res[-1] = np.append(res[-1],new_list1[p].hour) #убрал час начала сессии
            if (8<new_list1[p].hour<18): #добавил время суток(0-ночь, 1-день, 2-вечер)
                res[-1] = np.append(res[-1],1)
            elif new_list1[p].hour>=18:
                res[-1] = np.append(res[-1],2)
            else:
                res[-1] = np.append(res[-1],0)
            #res[-1] = np.append(res[-1],new_list1[p].weekday()) - #убрал день недели
            if new_list1[p].weekday()>4:
                res[-1] = np.append(res[-1],0)#добавил индикатор рабочего дня (0-выходной, 1-рабочий)
            else:
                res[-1] = np.append(res[-1],1)
            res[-1] = np.append(res[-1],season_dic[new_list1[p].month])#добавил сезон(0-зима,1-весна,2-лето,3-осень)
            res[-1] = np.append(res[-1],int(i[-8:-4]))
    col = []
    for i in range(session_length):
            col.append('site'+str(i+1))
    col.append('session_timespan')
    col.append('#unique_sites')
    col.append('top30_sites_visiting')
    #col.append('start_hour')
    col.append('daytime')
    #col.append('day_of_week')
    col.append('workday')
    col.append('season')
    col.append('target')
    session_data = pd.DataFrame(res, columns = col,  dtype=int)
    return session_data
```
