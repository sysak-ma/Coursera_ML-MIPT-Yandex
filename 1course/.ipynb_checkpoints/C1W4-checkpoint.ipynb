Импорт библиотек

```python

import numpy as np
import pandas as pd
import scipy
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

Создание выборки объема size
Распределение хи-квадрат

$$
F(x) = \frac{0.5^{\frac{k}{2}}}{Г(\frac{k}{2})}x^{\frac{k}{2}-1}e^{-\frac{x}{2}}
$$

, где Г - гамма-распределение

```python
k = 2
size = 1000
chi2 = scipy.stats.chi2(k)
sample = chi2.rvs(size)
```

Гистограмма с теоретической плотностью распределения $x^2$

```python
plt.figure(figsize=(10,6))

df = pd.DataFrame(sample, columns=['KDE'])
plt.hist(sample, bins=50, density=True)

x = np.linspace(-5, 18, 1000)
pdf = chi2.pdf(x)
plt.plot(x, pdf, label='theoretocal pdf', alpha=0.5, c='r')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Мат ожидание
$E[Y] = k$

Дисперсия
$D[Y] = 2k$

$\overline{X_n} \approx \sim N(k,\frac{2k}{n})$

$\overline{X_5} \approx \sim N(2,\frac{4}{5})$

```python
n = 5

means = []

for i in np.arange(1000):
    x = chi2.rvs(n)
    means.append(np.mean(x))
    
means = np.array(means)
mx = means.max()
# Параметры аппрооксимирующего нормального распределения
approx_mean = k
approx_std = np.sqrt(2. * k / n)
norm_rv = scipy.stats.norm(approx_mean, approx_std)
# График нормального распределения
x_norm = np.linspace(0, mx, 100)
pdf = norm_rv.pdf(x_norm)
plt.plot(x_norm, pdf, label='Normal distribution')

plt.hist(means, bins=30, density=True, label='Mean values dist')

plt.legend()
plt.xlabel('Mean')
plt.ylabel('Probability')
```
