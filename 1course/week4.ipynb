````python
# подключаем библиотеки
from matplotlib import pylab as plt
%matplotlib inline
import numpy as np
import scipy.stats as sts
````

Полукруглое распределение: $$f(x) = \dfrac{2}{\pi} \cdot \sqrt{1 - x^2}$$

````python
# создаем случайную величину и генерируем выборку размера 1000
rv = sts.semicircular()
sample = rv.rvs(1000)
````

````python
# строим гистограмму выборки и плотность распределения 
x = np.arange(-1, 1.01, 0.01)
y = rv.pdf(x)

plt.hist(sample, density = True)
plt.plot(x, y)

plt.xlabel('$x$')
plt.ylabel('$f(x)$')

plt.show()
````

Всюду далее пользуемся формулой для искомого нормального распределения: если матожидание нашей случайной величины есть $E$ (можно получить функцией ```rv.mean()```), а стандартное отклонение $\sigma$ (```rv.std()```), выборка имеет размер $n$, то нас интересует нормальное распределение с параметрами $E$ и $\sigma / \sqrt{n}$.

$n = 5$

````python
# создаем выборку и выводим ее часть на экран
n = 5
sample = []
for _ in range(1000):
    sample.append(rv.rvs(n).mean())
    
sample[:10]
````

````python
# строим гистограмму и плотность распределения поверх нее
plt.hist(sample, density = True)

approxDistr = sts.norm(rv.mean(), rv.std() / np.sqrt(n))
x = np.arange(-1, 1.01, 0.01)
y = approxDistr.pdf(x)
plt.plot(x, y)

plt.xlabel('$x$')
plt.ylabel('$f(x)$')

plt.show()
````
